```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
location="datasets/DC_Properties.csv"
df=pd.read_csv(location)
df.head(20)
```

```python
df.columns
```

```python
df.rename(columns={'Unnamed: 0': 'Number'}, inplace =True)
df.head()
```

```python
df.set_index('Number', inplace=True)
df.head()
```

```python
stats_df = df.drop(['NUM_UNITS','AYB','YR_RMDL','STORIES','SALEDATE','QUALIFIED','SALE_NUM','GBA','BLDG_NUM','STYLE','STRUCT','GRADE','EXTWALL','INTWALL','USECODE','SOURCE','CMPLX_NUM',"FULLADDRESS",'CITY','STATE','ZIPCODE','NATIONALGRID','LATITUDE', 'LONGITUDE', 'ASSESSMENT_NBHD','ASSESSMENT_SUBNBHD','CENSUS_TRACT','CENSUS_BLOCK','WARD','CMPLX_NUM','LIVING_GBA','SQUARE','X','Y','QUADRANT'], axis=1)
stats_df.head(20)
```

```python
df_no_missing = stats_df.dropna()
df_no_missing.head(20)
```

```python
#add more details to scatterplot
sns.lmplot(x='BEDRM', y='EYB', data=stats_df, 
           fit_reg=False) #remove regression line
```

```python
plt.subplots(figsize=(15,8))
sns.boxplot(data=stats_df)
```

```python
corr = stats_df.corr()
plt.subplots(figsize=(15,8))
sns.heatmap(corr, vmin=-1, annot=True)
```
